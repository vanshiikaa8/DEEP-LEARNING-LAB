```python
# Linear Regression using Keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
# Data
X = np.linspace(0, 10, 100)
Y = 3 * X + 2 + np.random.randn(*X.shape) * 0.5
# Build model
model = Sequential([
    Dense(1, input_shape=(1,), activation='linear')
])
# Compile model
model.compile(optimizer='sgd', loss='mse')
# Train model
history = model.fit(X, Y, epochs=100, verbose=0)
# Get learned parameters
weights, bias = model.layers[0].get_weights()
print(f"Trained Weight: {weights[0][0]:.4f}, Bias: {bias[0]:.4f}")
# Predictions
Y_pred = model.predict(X)
# Visualization
plt.scatter(X, Y, label='Original Data')
plt.plot(X, Y_pred, color='red', label='Fitted Line')
plt.legend()
plt.title('Keras Linear Regression')
plt.show()
```

```python
# Linear Regression using TensorFlow
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
# Generate synthetic data
X = np.linspace(0, 10, 100)
Y = 3 * X + 2 + np.random.randn(*X.shape) * 0.5  # Add some noise
# Define variables for weights and bias
W = tf.Variable(np.random.randn(), name="weight", dtype=tf.float32)
b = tf.Variable(np.random.randn(), name="bias", dtype=tf.float32)
# Learning rate and epochs
learning_rate = 0.01
epochs = 100
# Training loop
for epoch in range(epochs):
    with tf.GradientTape() as tape:
        Y_pred = W * X + b
        loss = tf.reduce_mean(tf.square(Y - Y_pred))
    gradients = tape.gradient(loss, [W, b])
    W.assign_sub(learning_rate * gradients[0])
    b.assign_sub(learning_rate * gradients[1])

    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1}: Loss = {loss.numpy():.4f}, W = {W.numpy():.4f}, b = {b.numpy():.4f}")
# Final prediction
Y_pred = W * X + b
# Visualization
plt.scatter(X, Y, label='Original Data')
plt.plot(X, Y_pred, color='red', label='Fitted Line')
plt.legend()
plt.title('TensorFlow Linear Regression')
plt.show()
```

```python
# Linear Regression using PyTorch

import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Generate data
X = torch.linspace(0, 10, 100).unsqueeze(1)
Y = 3 * X + 2 + torch.randn(X.size()) * 0.5

# Define model
model = nn.Linear(1, 1)

# Define loss and optimizer
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# Training loop
epochs = 100
for epoch in range(epochs):
    # Forward pass
    Y_pred = model(X)
    loss = criterion(Y_pred, Y)

    # Backward pass
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 10 == 0:
        [w, b] = model.parameters()
        print(f"Epoch {epoch+1}: Loss = {loss.item():.4f}, W = {w.item():.4f}, b = {b.item():.4f}")

# Prediction
Y_pred = model(X).detach()

# Visualization
plt.scatter(X, Y, label='Original Data')
plt.plot(X, Y_pred, color='red', label='Fitted Line')
plt.legend()
plt.title('PyTorch Linear Regression')
plt.show()
```
